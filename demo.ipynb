```python
from configparser import ConfigParser

class Config:
    def __init__(self, config_file=r"D:\python scripts\Generative AI\Langchain\Udemy langraph\AgenticAiworkspace\AgenticChatbotProject\src\langgraphagenticai\ui\uiconfigfile.ini"):
        self.config = ConfigParser()
        self.config.read(config_file)

    def get_llm_options(self):
        return self.config['DEFAULT'].get('LLM_OPTIONS').split(', ')

    def get_usecase_options(self):
        return self.config['DEFAULT'].get('USECASE_OPTIONS').split(', ')

    def get_groq_model_options(self):
        return self.config['DEFAULT'].get('GROQ_MODEL_OPTIONS').split(', ')

    def get_page_title(self):
        return self.config['DEFAULT'].get('PAGE_TITLE')


config = Config()
LLM_OPTIONS = config.get_llm_options()
USECASE_OPTIONS = config.get_usecase_options()
GROQ_MODEL_OPTIONS = config.get_groq_model_options()
PAGE_TITLE = config.get_page_title()

```

```python
print("LLM Options:", LLM_OPTIONS)
print("Usecase Options:", USECASE_OPTIONS)
print("GROQ Model Options:", GROQ_MODEL_OPTIONS)
print("Page Title:", PAGE_TITLE)
```
